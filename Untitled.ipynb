```python
mean1 = df['bill_length_mm'].mean()
mean2 = df['bill_depth_mm'].mean()
mean3 = df['flipper_length_mm'].mean()
mean4 = df['bill_depth_mm'].mean()
df['bill_length_mm']=df['bill_length_mm'].fillna(mean1)
df['bill_depth_mm']=df['bill_depth_mm'].fillna(mean2)
df['flipper_length_mm']=df['flipper_length_mm'].fillna(mean3)
df['body_mass_g']=df['body_mass_g'].fillna(mean4)

```

```python
df.dropna(subset=['sex'],inplace=True)
```

```python
df.isnull().sum()
```

```python
num = df.select_dtypes(include=['float64', 'int64']).columns
print("Numerical columns:", num)
```

```python
def remove_outliers(df, columns):
    for i in columns:
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df = df[(df[i] < lower) | (df[i] > upper)]
        return df

df_clean=remove_outliers(df, num)
```

```python
df['sex'] = df['sex'].str.lower()
```

## Bivariate analysis

```python
print(df[num].corr())
```

```python
for i, col1 in enumerate(num):
        for j, col2 in enumerate(num):
            if i < j: 
                plt.figure(figsize=(10, 6))
                sns.scatterplot(x=df[col1], y=df[col2])
                plt.title(f'Scatter Plot of {col1} vs {col2}')
                plt.xlabel(col1)
                plt.ylabel(col2)
                plt.show()
```

```python
def plot_categorical_numerical(df):
    categorical = df.select_dtypes(include=['object']).columns
    num = df.select_dtypes(include=['float64', 'int64']).columns
    
    for cat in categorical:
        for nx in num:
            plt.figure(figsize=(12, 6))
            sns.barplot(x=cat, y=nx, data=df)
            plt.title(f'Bar Plot of {cat} vs {nx}')
            plt.xlabel(cat)
            plt.ylabel(nx)
            plt.show()
            
            plt.figure(figsize=(12, 6))
            sns.violinplot(x=cat, y=nx, data=df)
            plt.title(f'Violin Plot of {cat} vs {nx}')
            plt.xlabel(cat)
            plt.ylabel(nx)
            plt.show()
            
            plt.figure(figsize=(12, 6))
            sns.boxplot(x=cat, y=nx, data=df)
            plt.title(f'Box Plot of {cat} vs {nx}')
            plt.xlabel(cat)
            plt.ylabel(nx)
            plt.show()

plot_categorical_numerical(df)
```
